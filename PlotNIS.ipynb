```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

Read csv file generated by UKF run containing NIS values for laser and radar measurement. Separate them by sensor type.

```python
df = pd.read_csv('./build/nis_out.csv', index_col=0)
laser_nis = df[df['Sensor Type (L/R)'] == 'L']['NIS']
radar_nis = df[df['Sensor Type (L/R)'] == 'R']['NIS']
```

What does the 95th percentile NIS value look like?

```python
print("Laser NIS 95th percentile: %.2f" % laser_nis.quantile(0.95))
print("Radar NIS 95th percentile: %.2f" % radar_nis.quantile(0.95))
```

Now plot NIS values

```python
plt.figure(figsize=(15,5))

# Plot Lidar NIS
plt.subplot(121)
plt.plot(laser_nis)
plt.title('Lidar')
plt.ylabel('NIS')
plt.xlabel('Timestamp (us)')
plt.ylim([0, 20]);
plt.axhline(y=7.8, linewidth=1, color='r', label='NIS=7.8');
plt.legend()


# Plot Radar NIS
plt.subplot(122)
plt.plot(radar_nis)
plt.title('Radar')
plt.ylabel('NIS')
plt.xlabel('Timestamp (us)')
plt.ylim([0, 20]);
plt.axhline(y=7.8, linewidth=1, color='r', label='NIS=7.8');
plt.legend()
```
